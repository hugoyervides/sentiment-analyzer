```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

NA = open("twitter-north-america.txt","r")
SA = open("twitter-south-america.txt","r")
Asia = open("twitter-asia.txt","r")
ME = open("twitter-middle_east.txt","r")
CA = open("twitter-central-america.txt","r")
AFRICA = open("twitter-africa.txt","r")
EUROPE = open("twitter-europe.txt","r")



NA = list(NA)
NAnp = np.array(NA)
NA_positive = sum((NAnp=='pos\n')*1)
NA_negative = sum((NAnp=='neg\n')*1)

SA = list(SA)
SAnp = np.array(SA)
SA_positive = sum((SAnp=='pos\n')*1)
SA_negative = sum((SAnp=='neg\n')*1)

Asia = list(Asia)
Asianp = np.array(Asia)
Asia_positive = sum((Asianp=='pos\n')*1)
Asia_negative = sum((Asianp=='neg\n')*1)


ME = list(ME)
MEnp = np.array(ME)
ME_positive = sum((MEnp=='pos\n')*1)
ME_negative = sum((MEnp=='neg\n')*1)

CA = list(CA)
CAnp = np.array(CA)
CA_positive = sum((CAnp=='pos\n')*1)
CA_negative = sum((CAnp=='neg\n')*1)

AFRICA = list(AFRICA)
AFRICAnp = np.array(AFRICA)
AFRICA_positive = sum((AFRICAnp=='pos\n')*1)
AFRICA_negative = sum((AFRICAnp=='neg\n')*1)

EUROPE = list(EUROPE)
EUROPEnp = np.array(EUROPE)
EUROPE_positive = sum((EUROPEnp=='pos\n')*1)
EUROPE_negative = sum((EUROPEnp=='neg\n')*1)



continent=['N-A','S-A','Asia','M-E','C-A','Africa','Europe']
Positive=[NA_positive,SA_positive,Asia_positive,ME_positive,CA_positive,AFRICA_positive,EUROPE_positive]
Negative=[NA_negative,SA_negative,Asia_negative,ME_negative,CA_negative,AFRICA_negative,EUROPE_negative]
xpos = np.arange(len(continent))

plt.bar(xpos-0.2,Positive,label='positive',width=0.4)
plt.bar(xpos+0.2,Negative,label='negative',width=0.4)
plt.xticks(xpos,continent)
plt.ylabel('Number of tweets')
plt.title('Sentiment analysis in different continents/areas')
plt.legend()
```

```python
Pos = np.array(Positive)
Neg = np.array(Negative)

Porcentaje = Pos*100/(Pos+Neg)

plt.bar(xpos,Porcentaje,width=0.4)
plt.xticks(xpos,continent)
plt.ylabel('Percentage')
plt.title('Percentage of positive tweets per continent/area')
```
